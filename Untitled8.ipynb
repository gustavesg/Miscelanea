```python
#IMPORTA BIBLIOTECAS
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#IMPORTA BASE
df = pd.read_csv('/content/drive/MyDrive/2023 QS World University Rankings.csv')
```

```python
df[df['location'] == 'Brazil'].head(20)
```

```python
#DATA FRAME APENAS DAS 10 PRIMEIRAS INSTITUIÇÕES DO BRASIL
df_brazil = df[df['location'] == 'Brazil'].head(10)

#TRANSFORMA COLUNA DE ACADEMIC REPUTATION SCORE PRO TIPO FLOAT
df_brazil['ar score'] = pd.to_numeric(df_brazil['ar score'])

#ORDENA DATA FRAME
df_brazil = df_brazil.sort_values(by = 'ar score')

############### GRÁFICO #################

#EIXO
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

#GRÁFICO BARRAS HORIZONTAIS
eixo.barh(df_brazil['institution'],df_brazil['ar score'], edgecolor = 'k')

#TITLE PRINCIPAL E EIXO X
eixo.set_title('Academic Reputation Score das Top 10 do Brasil')
eixo.set_xlabel('score')


```

```python
#DATA FRAME APENAS DAS 10 PRIMEIRAS INSTITUIÇÕES DO BRASIL
df_brazil = df[df['location'] == 'Brazil'].head(10)

#TRANSFORMA COLUNA DE ACADEMIC REPUTATION SCORE PRO TIPO FLOAT
df_brazil['er score'] = pd.to_numeric(df_brazil['er score'])

#ORDENA DATA FRAME
df_brazil = df_brazil.sort_values(by = 'er score')

############### GRÁFICO #################

#EIXO
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

#GRÁFICO BARRAS HORIZONTAIS
eixo.barh(df_brazil['institution'],df_brazil['er score'], color = 'y',edgecolor = 'k')

#TITLE PRINCIPAL E EIXO X
eixo.set_title('Employer Reputation Score das Top 10 do Brasil')
eixo.set_xlabel('score')
```

```python
#DATA FRAME APENAS DAS 10 PRIMEIRAS INSTITUIÇÕES DO BRASIL
df_brazil = df[df['location'] == 'Brazil'].head(10)

#TRANSFORMA COLUNA DE ACADEMIC REPUTATION SCORE PRO TIPO FLOAT
df_brazil['fsr score'] = pd.to_numeric(df_brazil['fsr score'])

#ORDENA DATA FRAME
df_brazil = df_brazil.sort_values(by = 'fsr score')

############### GRÁFICO #################

#EIXO
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

#GRÁFICO BARRAS HORIZONTAIS
eixo.barh(df_brazil['institution'],df_brazil['fsr score'], color = 'orange',edgecolor = 'k')

#TITLE PRINCIPAL E EIXO X
eixo.set_title('Faculty Student Score das Top 10 do Brasil')
eixo.set_xlabel('score')
```

```python
#DATA FRAME TOP10 SEGUNDO AR SCORE
df_top_10 = df.sort_values('ar score', ascending = False)[0:20]
#CRIA FIG
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])
#GRAFICO DE BARRAS HORIZONTAIS
eixo.barh(df_top_10['institution'], df_top_10['ar score'], edgecolor = 'k')
#MÉDIA GERAL DE ARSCORE
media_geral = df['ar score'].mean()
media_top_10 = df_top_10['ar score'].mean()
#LINHA VERTICAL PRA MÉDIA GERAL
eixo.axvline(media_geral,color = 'k',linestyle = '--')
#LINHA VERTICAL PRA MÉDIA DO TOP 10
eixo.axvline(media_top_10,color = 'r',linestyle ='--')
#TITULO
eixo.set_title('Top 20 World by Academic Reputation Score')

fig.show()
```

```python
#LISTA DOS PAISES COM MELHORES FACULDADES
paises_melhores = ['US','UK','CN','JP','RU','DE','IT','IN','KR','AU']

#LISTA PARA AS MÉDIAS DE AR SCORE NOS PAISES COM AS MELHORES FACULDADES
lista_medias = []

for paises in paises_melhores:

  lista_medias.append(df[df['location code'] == f'{paises}']['ar score'].mean())

#CRIA FIGURE
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.bar(paises_melhores,sorted(lista_medias),edgecolor = 'k')
eixo.set_title('Média de AR score nos países com faculdades no TOP20')
fig.show()
```

```python
#DATAFRAMES 50 PIORES E 50 MELHORES 
melhores = df[0:50]
piores = df[-50:]

#CRIA FIGURE
fig = plt.figure(figsize = (5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(melhores['ar score'], melhores['er score'], color = 'green')
eixo.scatter(piores['fsr score'], melhores['fsr score'], color = 'red')

eixo.legend(['50 melhores','50 piores'], loc = 'lower right')

eixo.set_xlabel('AR score')
eixo.set_ylabel('ER score')

eixo.annotate('Média ER score', xy = (40,df['er score'].mean()), xytext = (30,df['er score'].mean()+20), arrowprops = dict(facecolor = 'k',width = 0.5, headwidth = 6))
eixo.annotate('Média AR score', xy = (df['er score'].mean(),60), xytext = (df['er score'].mean()+20,59), arrowprops = dict(facecolor = 'k',width = 0.5, headwidth = 6))
eixo.axhline(df['er score'].mean(), linestyle = '--', color = 'k')
eixo.axvline(df['ar score'].mean(), linestyle = '--', color = 'k')
```
